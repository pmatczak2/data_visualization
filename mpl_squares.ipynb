```python
import matplotlib.pyplot as plt

squares = [1, 4, 9, 16, 25]
fig, ax = plt.subplots()
ax.plot(squares, linewidth=3)

ax.set_title("Square Numbers", fontsize=24)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=14)

ax.tick_params(axis='both', labelsize=14)

plt.show()
```

#                               plotting a siries of points with scatter()

```python
import matplotlib.pyplot as plt

x_value = [1, 2, 3, 4, 5]#this contains the numbers to be squared
y_value = [1, 4, 9, 16, 25]#this contains the square of each number

plt.style.use('seaborn')#gave the graph more style
fig, ax = plt.subplots()
ax.scatter(x_value, y_value, s=100)#when the x,y lists are passed to scatter() Matplotlib reads one value,
                                   #then plots each point.

ax.set_title("Square Numbers", fontsize=24)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=14)

ax.tick_params(axis='both', labelsize=14)

plt.show()
```

# calculating data automatically

```python
import matplotlib.pyplot as plt

x_values = range(1, 1001)#start the x_values contatining number 1 - 1000
y_values = [x**2 for x in x_values]# list comprehension loops generates y-value by looping through the x-values
                           #(for x in x_values) squaring eaxh number (x**2) and storing the results in y_values

plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, s=10)#pass the input and output lists to scatter()

ax.set_title("Square Numbers", fontsize=24)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=14)

ax.tick_params(axis='both', labelsize=14)

ax.axis([0, 1100, 0, 1100000])#the axis() method rquires four values, min, and max for x-axis and y-axis 

plt.show()
```

# using color map

```python
import matplotlib.pyplot as plt

x_values = range(1, 1001)
y_values =  [x**2 for x in x_values]
                           
plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, c=y_values, cmap=plt.cm.Blues, s=10)#pass the y-values to c, tell python colormap
#to use using the cmap argument. this code colors the points with lower y-values light blue and color the points 
#with higher y-value dark blue

ax.set_title("Square Numbers", fontsize=24)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=14)

ax.tick_params(axis='both', labelsize=14)

ax.axis([0, 1100, 0, 1100000])
plt.show()
```

# try it yourself 15-1 cubes

```python
import matplotlib.pyplot as plt

x_values = range(1, 5001)
y_values = [x**3 for x in x_values]

plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, c=y_values, cmap=plt.cm.Greens, s=10)

ax.set_title("Square Numbers", fontsize=24)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=14)

ax.tick_params(axis='both', labelsize=14)

ax.axis([0, 5100, 0, 130000000000])

plt.show()
```
